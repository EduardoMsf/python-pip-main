```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

```python
##Histogram
plt.hist(data, bins=10, histtype='barstacked', alpha=0.7, color='b', edgecolor='black')
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showmeans=True, showcaps=True, showbox=True, showfliers=True)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
