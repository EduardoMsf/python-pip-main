```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python

tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='viridis')
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='stack', palette='viridis')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, palette='viridis', bw_adjust=0.5)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack', bins=15, palette='viridis')
plt.show()
```
