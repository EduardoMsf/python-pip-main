```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(country, population, color=['blue', 'red'], width=0.5)
plt.xticks(np.arange(len(country)), ('Indie', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

```python
plt.barh(country, population, color=['blue', 'red'])
plt.yticks(np.arange(len(country)), ('Indie', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```
