```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,10))
axes[0].plot(x, y, 'r', label='$sin(x)$')
axes[0].set_title('Relacion entre X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y, x, 'b')
axes[1].set_title('Relacion entre Y - X')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,10))
axes[0].plot(x, y, 'r', label='$sin(x)$')
axes[0].set_title('Relacion entre X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y, x, 'b')
axes[1].set_title('Relacion entre Y - X')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```
