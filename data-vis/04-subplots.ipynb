```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(np.sin(x), 'r')
axes[0,1].plot(np.cos(y) ,'b')
axes[0,2].plot(np.tan(y) ,'y')
axes[0,3].plot(np.cos(y)**2 ,'purple')
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2,4)
ax1.plot(np.sin(x), 'r')
ax2.plot(np.cos(y) ,'b')
ax3.plot(np.tan(y) ,'y')
ax4.plot(np.cos(y)**2 ,'purple')
fig.tight_layout()
```
