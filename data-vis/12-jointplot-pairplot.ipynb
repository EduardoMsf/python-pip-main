```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='kde', marginal_ticks=True,marginal_kws=dict( fill=False, multiple='stack'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='viridis', corner=True, kind='scatter')
plt.show()
```
