```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
markers = {'Lunch': 'D', 'Dinner': 's'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.lineplot(x='total_bill', y='tip', data=tips, hue='day')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='day', style='time', size='size', col='time')
plt.show()
```
